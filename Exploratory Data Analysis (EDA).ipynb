```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv("train_and_test.csv")
```

```python
df.head()
```

```python
#shape descibes 
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# checking missing values
df.isnull().sum()
```

```python
# for removing null valuee 
df.dropna(inplace = True)
```

```python
#Remove Unnecessary Columns (Many zero columns)
zero_cols = [col for col in df.columns if 'zero' in col.lower()]
df.drop(columns=zero_cols, inplace=True)

df.head()
```

```python
# rename  target coloumn
df.rename(columns={'2urvived': 'Survived'}, inplace=True) 
```

```python
#Data Visualization (MOST IMPORTANT)
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

```python
#Gender vs Survival

#(Sex: 0 = Female, 1 = Male)
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Sex vs Survival")
plt.show() 
```

```python
#Passenger Class vs Survival
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Passenger Class vs Survival")
plt.show()
```

```python
#age distribution 
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=30)
plt.title("Age Distribution")
plt.show()
```

```python
# FARE DISTRIBUTIOIN
plt.figure(figsize=(8,5))
sns.histplot(df['Fare'], bins=30)
plt.title("Fare Distribution")
plt.show()
```

```python
# CORELATION HEATMAP 
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
OBSERVATIONS:
1. Females survived more than males.
2. Passengers in Pclass 1 had higher survival rate.
3. Younger passengers had better survival chances.
4. Fare and Pclass show correlation with survival.
```
